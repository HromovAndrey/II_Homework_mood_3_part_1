```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/HalyshAnton/IT-Step-Pyton-AI/main/module3/data/insurance.csv")
df = df.select_dtypes('number')
```

# Завдання 2
Розділіть дані на залежну y та незалежну X ознаки. Розділіть дані на тренувальну та тестову частини

```python
from sklearn.model_selection import train_test_split

y = df['expenses']
X = df.drop(columns='expenses')

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)

print("Розмір тренувальної вибірки:", X_train.shape)
print("Розмір тестової вибірки:", X_test.shape)
```

# Завдання 3
Натренуйте модель та виведіть такі метрики: MAE, RMSE, R2

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")
print(f"R2: {r2}")
```

# Завдання 4
Напишіть рівняння регресії за прикладом:
Ціна = 1.02*Температура - 15.6

```python
coefficients = model.coef_
intercept = model.intercept_
regression_equation = "Expenses = " + " + ".join([f"{coef:.2f}*{col}" for coef, col in zip(coefficients, X.columns)]) + f" + {intercept:.2f}"

print("Рівняння регресії:")
print(regression_equation)
```

# Завдання 5 (оптимально)
Перевірте як видалення викидів вплине на результат

```python
def remove_outliers_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    return df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

df_no_outliers = remove_outliers_iqr(df)
y_no_outliers = df_no_outliers['expenses']
X_no_outliers = df_no_outliers.drop(columns='expenses')

X_train_no_outliers, X_test_no_outliers, y_train_no_outliers, y_test_no_outliers = train_test_split(X_no_outliers, y_no_outliers, train_size=0.8, random_state=42)
model_no_outliers = LinearRegression()
model_no_outliers.fit(X_train_no_outliers, y_train_no_outliers)
y_pred_no_outliers = model_no_outliers.predict(X_test_no_outliers)
mae_no_outliers = mean_absolute_error(y_test_no_outliers, y_pred_no_outliers)
rmse_no_outliers = np.sqrt(mean_squared_error(y_test_no_outliers, y_pred_no_outliers))
r2_no_outliers = r2_score(y_test_no_outliers, y_pred_no_outliers)

print(f"MAE без викидів: {mae_no_outliers}")
print(f"RMSE без викидів: {rmse_no_outliers}")
print(f"R2 без викидів: {r2_no_outliers}")
```
